```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
image_path = 'Frozen Rose.jpg'
image = cv2.imread(image_path)
cv2_imshow(image)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
laplacian_abs = cv2.convertScaleAbs(laplacian)
cv2_imshow(laplacian_abs)
```

```python
if len(laplacian_abs.shape) == 2:
    laplacian_abs = cv2.cvtColor(laplacian_abs, cv2.COLOR_GRAY2BGR)
```

```python
alpha = 0.5
blended_image = cv2.addWeighted(image, alpha, laplacian_abs, 1 - alpha, 0)

# Display the blended image
cv2_imshow(blended_image)
```
