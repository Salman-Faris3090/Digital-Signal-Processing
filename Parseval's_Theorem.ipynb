```python
import numpy as np
import matplotlib.pyplot as plt

# Random signal
np.random.seed(0)
signal_length = 1000
signal = np.random.randn(signal_length)

# Computing Fourier transform
signal_fft = np.fft.fft(signal)

# Computing energy in spatial domain
energy_spatial = np.sum(np.abs(signal) ** 2) / signal_length

# Computing energy in frequency domain
energy_freq = np.sum(np.abs(signal_fft) ** 2) / signal_length

print("Energy in spatial domain:", energy_spatial)
print("Energy in frequency domain:", energy_freq)

# Plot the signal and its Fourier transform
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(signal)
plt.title("Original Signal")
plt.subplot(2, 1, 2)
plt.plot(np.abs(signal_fft))
plt.title("Fourier Transform of Signal")
plt.show()
```
